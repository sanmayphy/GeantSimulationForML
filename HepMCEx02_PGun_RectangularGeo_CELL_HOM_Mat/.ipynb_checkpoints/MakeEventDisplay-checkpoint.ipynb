```python
def DrawTrajectory(evt_i, traj_i) : 
    
     
    
    
    import matplotlib.pyplot as plt
    import matplotlib as mpl
    from mpl_toolkits.mplot3d import Axes3D
    
    plt.rcParams["figure.figsize"]=10,10
    
    nTraj = len(f['EventTree'].array('Trajectory_NStep')[evt_i])
    print('NTraj in evt ' + str(evt_i+1) + ' : ' + str(nTraj)  )
    Traj_NStep = int(f['EventTree'].array('Trajectory_NStep')[evt_i][traj_i])
    print('Traj ' + str(traj_i+1) + ' has Step: ' + str(Traj_NStep)  )
    
    #Traj_NStep = np.min(500, Traj_NStep)
    
    InitX = Trajectory_Step_InitPosX[evt_i][traj_i][0 : Traj_NStep]
    InitY = Trajectory_Step_InitPosY[evt_i][traj_i][0 : Traj_NStep]
    InitZ = Trajectory_Step_InitPosZ[evt_i][traj_i][0 : Traj_NStep]
    
    FinalX = Trajectory_Step_FinalPosX[evt_i][traj_i][0 : Traj_NStep]
    FinalY = Trajectory_Step_FinalPosY[evt_i][traj_i][0 : Traj_NStep]
    FinalZ = Trajectory_Step_FinalPosZ[evt_i][traj_i][0 : Traj_NStep]
    
    minX = np.min( [np.min(InitX), np.min(FinalX)] ) 
    maxX = np.max( [np.max(InitX), np.max(FinalX)] )
    
    minY = np.min( [np.min(InitY), np.min(FinalY)] ) 
    maxY = np.max( [np.max(InitY), np.max(FinalY)] )
    
    minZ = np.min( [np.min(InitZ), np.min(FinalZ)] ) 
    maxZ = np.max( [np.max(InitZ), np.max(FinalZ)] )

    mpl.rcParams['legend.fontsize'] = 10

    fig = plt.figure()
    ax = fig.gca(projection='3d')
    ax.set_xlim(minX, maxX)
    ax.set_ylim(minY, maxY)
    ax.set_zlim(minZ, maxZ)
    

    ListX, ListY, ListZ = [], [], []
    for step_i in range(Traj_NStep) : 
        listX = np.array([ InitX[step_i], FinalX[step_i] ])
        listY = np.array([ InitY[step_i], FinalY[step_i] ])
        listZ = np.array([ InitZ[step_i], FinalZ[step_i] ])
        
        ax.plot(listX, listY, listZ, label='Step', color='red')
        
        ListX.append(listX)
        ListY.append(listY)
        ListZ.append(listZ)
        
    
#     mpl.rcParams['legend.fontsize'] = 10

#     fig = plt.figure()
#     ax = fig.gca(projection='3d')
    
#     ax.plot(ListX, ListY, ListZ, label='Step')
#     ax.legend()
    
    plt.show()
    
```

```python
def DrawEvent(evt_i, sample='') : 
    
    import matplotlib.pyplot as plt
    import matplotlib as mpl
    from mpl_toolkits.mplot3d import Axes3D
    
    plt.rcParams["figure.figsize"]=20,20
    
    fig = plt.figure()
    ax = fig.gca(projection='3d')
    ax.set_xlim(-8000, 8000)
    ax.set_ylim(-5000, 5000)
    ax.set_zlim(-5000, 5000)
    
    ax.set_xlabel("Z [mm]")
    ax.set_ylabel("Y [mm]")
    ax.set_zlabel("X [mm]")
    
    # Cylinder
    x=np.linspace(-4000, 4000, 100)
    z=np.linspace(-8000, 8000, 100)
    Xc, Zc=np.meshgrid(x, z)
    Yc = np.sqrt(4000**2-Xc**2)
    
    # Draw parameters
    rstride = 20
    cstride = 10
    ax.plot_surface(Zc, Yc, Xc, alpha=0.2, rstride=rstride, cstride=cstride)
    ax.plot_surface(Zc, -Yc, Xc, alpha=0.2, rstride=rstride, cstride=cstride)
    
    
    # Inner cylinder
    x_in=np.linspace(-2000, 2000, 100)
    z_in=np.linspace(-8000, 8000, 100)
    Xc_in, Zc_in=np.meshgrid(x_in, z_in)
    Yc_in = np.sqrt(2000**2-Xc_in**2)
    
    ax.plot_surface(Zc_in, Yc_in, Xc_in, alpha=0.2, rstride=rstride, cstride=cstride, color='magenta')
    ax.plot_surface(Zc_in, -Yc_in, Xc_in, alpha=0.2, rstride=rstride, cstride=cstride, color='magenta')
    
    # Outer cylinder
    x_out=np.linspace(-3000, 3000, 100)
    z_out=np.linspace(-8000, 8000, 100)
    Xc_out, Zc_out=np.meshgrid(x_out, z_out)
    Yc_out = np.sqrt(3000**2-Xc_out**2)
    
    ax.plot_surface(Zc_out, Yc_out, Xc_out, alpha=0.2, rstride=rstride, cstride=cstride, color='cyan')
    ax.plot_surface(Zc_out, -Yc_out, Xc_out, alpha=0.2, rstride=rstride, cstride=cstride, color='cyan')
    
    
    nTraj = len(Trajectory_NStep[evt_i])
    print('NTraj in evt ' + str(evt_i+1) + ' : ' + str(nTraj)  )
    
    for traj_i in range(nTraj) : 
        
        traj_ch = Trajectory_Charge[evt_i][traj_i]
        trajColor = ''
        if(traj_ch == 0.) : trajColor = 'green'
        else : trajColor = 'red'
        
        #print('Traj : ' + str(traj_i) + ', Charge = ' +str(traj_ch))
        
        Traj_NStep = int(Trajectory_NStep[evt_i][traj_i])
        if(Traj_NStep > 500) : Traj_NStep = 500
        
        InitX = Trajectory_Step_InitPosX[evt_i][traj_i][0 : Traj_NStep]
        InitY = Trajectory_Step_InitPosY[evt_i][traj_i][0 : Traj_NStep]
        InitZ = Trajectory_Step_InitPosZ[evt_i][traj_i][0 : Traj_NStep]

        FinalX = Trajectory_Step_FinalPosX[evt_i][traj_i][0 : Traj_NStep]
        FinalY = Trajectory_Step_FinalPosY[evt_i][traj_i][0 : Traj_NStep]
        FinalZ = Trajectory_Step_FinalPosZ[evt_i][traj_i][0 : Traj_NStep]

        
        
        for step_i in range(Traj_NStep) : 
            
            initR = np.sqrt( pow(InitX[step_i],2) + pow(InitY[step_i],2) )
            if(initR < 2000 and traj_ch == 0.) : continue
                
            listX = np.array([ InitX[step_i], FinalX[step_i] ])
            listY = np.array([ InitY[step_i], FinalY[step_i] ])
            listZ = np.array([ InitZ[step_i], FinalZ[step_i] ])

            ax.plot(listZ, listY, listX, label='Step', color=trajColor)
    
    
    plt.savefig('Figures/' + sample + 'Event' + str(evt_i) + '.pdf' )
    plt.savefig('Figures/' + sample + 'Event' + str(evt_i) + '.png' )
    
    
    
    
```
