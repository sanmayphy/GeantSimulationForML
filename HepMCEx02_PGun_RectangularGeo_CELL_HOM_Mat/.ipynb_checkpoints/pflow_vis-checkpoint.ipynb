```python
import matplotlib.pyplot as plt
import numpy as np
import glob
```

```python
data = np.load('event_0.npy')
```

```python
data.shape
```

```python

n_total = len(glob.glob('event_*'))
n_events = 10 


fig,ax = plt.subplots(n_events*2,6,figsize=(12,n_events*3))


for event_i in range(n_events):
    
    rand_event = np.random.randint(n_total)
    data = np.load('event_'+str(rand_event)+'.npy')
    
    for charged_or_neutral in [0,1]:
        for layer_i in range(6):

            img = data[charged_or_neutral][layer_i]
            img[img > 0] = 1

            if charged_or_neutral==0:
                cmap='OrRd'
            else:
                cmap='Greens'
            ax[2*event_i+charged_or_neutral][layer_i].imshow(img,cmap=cmap)

plt.tight_layout()
plt.savefig('test.pdf')
plt.show()
```
